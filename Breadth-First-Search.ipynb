```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque

def bfs_taiwan(graph, start, goal):
    queue = deque([start])
    visited = set()
    path = []

    while queue:
        node = queue.popleft()
        path.append(node)

        if node == goal:
            break

        if node not in visited:
            visited.add(node)
            queue.extend(graph[node])

    return path

# Define Taiwan City Graph
taiwan_graph = {
    'Taipei': ['Taoyuan', 'Hsinchu'],
    'Taoyuan': ['Taipei', 'Hsinchu', 'Taichung'],
    'Hsinchu': ['Taipei', 'Taoyuan', 'Taichung'],
    'Taichung': ['Taoyuan', 'Hsinchu', 'Tainan'],
    'Tainan': ['Taichung', 'Kaohsiung'],
    'Kaohsiung': ['Tainan']
}

# Convert to NetworkX Graph
G = nx.Graph(taiwan_graph)

# Run BFS
bfs_path = bfs_taiwan(taiwan_graph, 'Taipei', 'Kaohsiung')

# Draw Graph
pos = nx.spring_layout(G)
plt.figure(figsize=(6, 4))
nx.draw(G, pos, with_labels=True, node_color='lightgray', edge_color='gray', node_size=1000, font_size=10)

# Highlight BFS path
path_edges = list(zip(bfs_path, bfs_path[1:]))
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=2)
nx.draw_networkx_nodes(G, pos, nodelist=bfs_path, node_color='red')

plt.title("BFS in Taiwan Cities")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque

def bfs_path(graph, start, goal):
    queue = deque([(start, [start])])
    visited = set()

    while queue:
        node, path = queue.popleft()
        if node == goal:
            return path

        if node not in visited:
            visited.add(node)
            for neighbor in graph[node]:
                queue.append((neighbor, path + [neighbor]))

    return []

# Define Taiwan City Graph
taiwan_graph = {
    'Taipei': ['Taoyuan', 'Hsinchu'],
    'Taoyuan': ['Taipei', 'Hsinchu', 'Taichung'],
    'Hsinchu': ['Taipei', 'Taoyuan', 'Taichung'],
    'Taichung': ['Taoyuan', 'Hsinchu', 'Tainan'],
    'Tainan': ['Taichung', 'Kaohsiung'],
    'Kaohsiung': ['Tainan']
}

# Convert to NetworkX Graph
G = nx.Graph(taiwan_graph)

# Run BFS
bfs_result = bfs_path(taiwan_graph, 'Taipei', 'Kaohsiung')

# Draw Graph
pos = nx.spring_layout(G)
plt.figure(figsize=(6, 4))
nx.draw(G, pos, with_labels=True, node_color='lightgray', edge_color='gray', node_size=1000, font_size=10)

# Highlight BFS path
path_edges = list(zip(bfs_result, bfs_result[1:]))
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=2)
nx.draw_networkx_nodes(G, pos, nodelist=bfs_result, node_color='red')

plt.title("BFS - Best Path in Taiwan Cities")
plt.show()
```
